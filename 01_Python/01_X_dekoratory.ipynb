# Dekorátory
* Dekorátory jsou funkce, které modifikují chování jiné funkce nebo metody.
* Používají se k přidání funkcionality bez změny původního kódu.
* Hodí se pro: logování, měření času, kontrolu práv, cache, validace vstupu

```python
from functools import wraps

def my_decorator(func):
    @wraps(func)          # preserves __name__, __doc__ of the original function
    def wrapper(*args, **kwargs):
        print(f"Before function {func.__name__}")
        result = func(*args, **kwargs)
        print("After function")
        return result
    return wrapper
```

```python
@my_decorator
def do_this():
    print("Doing this")
```

```python
@my_decorator
def do_that():
    print("Doing that")
```

```python
do_this()
```

```python
do_that()
```

```python
do_this()
# without @wraps the name would show "wrapper" instead of "do_this"
print(do_this.__name__)   # do_this
```

## Praktický příklad: měření času

Jeden z nejčastějších důvodů pro použití dekorátoru — měření doby běhu funkce bez změny jejího kódu.

```python
import time
from functools import wraps

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - start
        print(f"{func.__name__} took {elapsed:.4f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(range(n))

slow_sum(10_000_000)
```

## Praktický příklad: logování

Dekorátor, který zaloguje volání funkce — vstupní argumenty a návratovou hodnotu.

```python
def log(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"[LOG] calling {func.__name__}({args}, {kwargs})")
        result = func(*args, **kwargs)
        print(f"[LOG] {func.__name__} returned {result}")
        return result
    return wrapper

@log
def add(a, b):
    return a + b

add(3, 5)
```

## Dekorátor s argumenty

Někdy chceme dekorátor parametrizovat — např. `@repeat(3)`. Přidá se ještě jedna vrstva funkce, která argument přijme.

```python
def repeat(n):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(n):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(3)
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
```
